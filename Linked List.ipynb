# 3. Linked List I. Singly Linked List

* Definition
    * A linked list is a sequential list of nodes that hold data which point to other nodes also conatining data. A singly linked list hold a reference to the next node. In implementation, always maintain a reference to the head of the linked list & a reference to the tail of the linked list for quick 

* Terminology
    * Head: first node in a  linked list
    * Tail: last node in a linked list
    * Pointer: reference to another node
    * Node: object containing data & pointer
* Compare with Array


## (1). Implementation

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        
class SLinkedList:
    def __init__(self):
        self.headval = None
        
    # print a linked list(traverse a linked list)
    def listprint(self):
        printval = self.headval
        while printval != None:
            print(printval.dataval)
            printval = printval.nextval
    
    # insert at beginning
    def at_beginning(self, new_data):
        new_node = Node(new_data)
        # update new node next val to existing node
        new_node.nextval = self.headval
        self.headval = new_node
        
    # insert at end
    def at_end(self, new_data):
        new_node = Node(new_data)
        if self.headval == None:
            self.headval = new_node
            return
        last_node = self.headval
        while(last_node.nextval):
            last_node = last_node.nextval
        last_node.nextval = new_node
    
    # insert between two nodes
    def in_between(self, in_between_first, new_data):
        # in_between_first represent the first node of the two nodes that will
        # be inserted in another node
        if in_between_first == None:
            print('mentioned node is missing')
            return
        new_node = Node(new_data)
        new_node.nextval = in_between_first.nextval
        in_between_first.nextval = new_node
    
    # remove a node
    def remove_node(self, remove_key):
        #start with the first node
        head_val = self.headval
        #when target value is the first node
        if head_val != None:
            if head_val.dataval == remove_key:
                #delete the first node directly
                self.headval = head_val.nextval
                #target been delete, let it equal to none
                target = None
                return
        while head_val != None:
            if head_val.dataval == remove_key:
                break
            #if head_val is not target node, then it becomes the previous node
            previous_node = head_val
            #target node becomes the next node
            head_val = head_val.nextval
        #now we have found the target node, change the pointer, and target node to none
        previous_node.nextval = head_val.nextval
        head_val = None
        
        if head_val == None:
            return
```

### merge 2 sorted linked list

#### (a) recursive method

```python
def MergeTwoList(head1, head2):
    temp = None
    if head1 is None:
        return head2
    elif head2 is None:
        return head1
    elif head1.dataval <= head2.dataval:
        temp = head1
        temp.nextval = MergeTwoList(head1.nextval, head2)
    else:
        temp = head2
        temp.nextval = MergeTwoList(head1, head2.nextval)
    
    return temp
```
