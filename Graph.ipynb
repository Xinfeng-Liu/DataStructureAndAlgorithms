# Graph

## (1). Definition
A graph is a pictorial representation of a set of objects where some pairs of objects are connected by links. The interconnected objects are represented by points termed as vertices, and links that connect the vertices are called edges.


## (2). Create a Graph

```python
graph = {
    "a": ['b', 'c'],
    "b": ['a', 'd'],
    "c": ['a', 'd'],
    "d": ['e'],
    "e": ['d']
}
print(graph)
```

## (3). Modify a Graph

```python
class graph:
    def __init__(self, gdict=None):
        if gdict == None:
            gdict = {}
        self.gdict = gdict
        
    # display vertices of a graph
    def getVertices(self):
        return list(self.gdict.keys())
    
    # display edges of a graph
    def getEdges(self):
        edges = []
        for vertices in self.gdict:
            for nxvertices in self.gdict[vertices]:
                if {vertices, nxvertices} not in edges:
                    edges.append({vertices, nxvertices})
        return edges
    
    # add vertex
    def addVertex(self, vrtx):
        if vrtx not in self.gdict:
            self.gdict[vrtx] = []
    
    # add edge
    def addEdge(self, edge):
        edge = set(edge)
        (vrtx1, vrtx2) = tuple(edge)
        if vrtx1 in self.gdict:
            self.gdict[vrtx1].append(vrtx2)
        else:
            self.gdict[vrtx1] = [vrtx2]
        
```

```python
graph_elements = { 
   "a" : ["b","c"],
   "b" : ["a", "d"],
   "c" : ["a", "d"],
   "d" : ["e"],
   "e" : ["d"]
}
g = graph(graph_elements)
g.addEdge({'a', 'e'})
print(g.getEdges())
```
